```python
# getting india's ID first, OpenAQ API has a hierarchical structure, so we need to get the country ID for India first, then use that ID to get the locations within India.
url = "https://api.openaq.org/v3/countries"
headers = {
    "X-API-Key": OPENAQ_API_KEY
}
params = {
    "limit": 100    }

response = requests.get(url, headers=headers, params=params)
countries_data = response.json()
countries_data.keys()
```

```python
countries = countries_data["results"]
for country in countries:
    if country["name"] == "India":
       print(country)
```

```python
# now getting india's locations and fidning delhi in them
url = "https://api.openaq.org/v3/locations"
#headers = {
 #   "X-API-Key": OPENAQ_API_KEY
  #  } 
params = {
    "countries_id": "9",
    "limit": 60 }

response = requests.get(url, headers=headers, params=params)
print(response.status_code)
print(response.url)
```

```python
data = response.json()
locations = data["results"]
```

```python
for loc in locations:
    print(loc["id"], "-", loc["name"], "-", loc.get("locality"), "-", loc["country"]["name"])
```

```python
# adding all delhi locations to one list
delhi_locations = []

for loc in locations:
    name = loc["name"].lower()
    locality = str(loc.get("locality")).lower()

    if "delhi" in name or "delhi" in locality:
        delhi_locations.append(loc)

len(delhi_locations)
```

```python
# print clearly:
for loc in delhi_locations:
    print(
        loc["id"],
        "-",
        loc["name"],
        "-",
        loc.get("locality"),
        "-",
        loc["coordinates"]
    )
```

```python
# converting to a df
rows = []

for loc in delhi_locations:
    row = {
        "id": loc["id"],
        "name": loc["name"],
        "locality": loc.get("locality"),
        "latitude": loc["coordinates"]["latitude"],
        "longitude": loc["coordinates"]["longitude"]
    }
    rows.append(row)

delhi_locs_df = pd.DataFrame(rows)
delhi_locs_df
```

```python
delhi_locs_df.to_csv("../data/raw/openaq_delhi_locs.csv", index=False)
```

```python
for loc in delhi_locations:
    print("\nLOCATION:", loc["id"], "-", loc["name"])
    
    for sensor in loc["sensors"]:
        print(
            "  sensor_id:",
            sensor["id"],
            "| parameter:",
            sensor["parameter"]["name"],
            "| display:",
            sensor["parameter"]["displayName"],
            "| units:",
            sensor["parameter"]["units"]
        )
```

```python
pm25_sensors = []

for loc in delhi_locations:
    for sensor in loc["sensors"]:
        if sensor["parameter"]["name"] == "pm25":
            pm25_sensors.append({
                "location_id": loc["id"],
                "location_name": loc["name"],
                "sensor_id": sensor["id"],
                "parameter": sensor["parameter"]["name"],
                "units": sensor["parameter"]["units"],
                "latitude": loc["coordinates"]["latitude"],
                "longitude": loc["coordinates"]["longitude"]
            })

pm25_sensors
```

```python
pm25_sensors_df = pd.DataFrame(pm25_sensors)
pm25_sensors_df
```

```python
sensor_id = pm25_sensors_df.iloc[0]["sensor_id"]
sensor_id
```

```python
url = f"https://api.openaq.org/v3/sensors/{sensor_id}/measurements"

params = {
    "limit": 100
}

response = requests.get(url, headers=headers, params=params)

print(response.status_code)
print(response.url)
```

```python
measurements_data = response.json()

measurements_data.keys()
```

```python
measurements = measurements_data["results"]

len(measurements)
```

```python
measurements[0]
```

```python
measurements[0].keys()
```

```python
from pprint import pprint

pprint(measurements[0])
```

```python
measurement_rows = []

for m in measurements:
    row = {
        "sensor_id": sensor_id,
        "value": m["value"],
        "parameter": m["parameter"]["name"],
        "unit": m["parameter"]["units"],
        "datetime_utc": m["period"]["datetimeFrom"]["utc"],
        "datetime_local": m["period"]["datetimeFrom"]["local"]
    }
    measurement_rows.append(row)

measurements_df = pd.DataFrame(measurement_rows)
measurements_df.head()
```

```python
measurements_df.shape
```

```python
sensor_info = pm25_sensors_df[pm25_sensors_df["sensor_id"] == sensor_id].iloc[0]

measurements_df["location_id"] = sensor_info["location_id"]
measurements_df["location_name"] = sensor_info["location_name"]
measurements_df["latitude"] = sensor_info["latitude"]
measurements_df["longitude"] = sensor_info["longitude"]

measurements_df.head()
```

```python
all_measurement_rows = []

for _, sensor_row in pm25_sensors_df.iterrows():
    sensor_id = sensor_row["sensor_id"]

    url = f"https://api.openaq.org/v3/sensors/{sensor_id}/measurements"
    params = {"limit": 100}

    response = requests.get(url, headers=headers, params=params)

    if response.status_code != 200:
        print("Failed:", sensor_id, response.status_code)
        continue

    measurements = response.json()["results"]

    for m in measurements:
        all_measurement_rows.append({
            "location_id": sensor_row["location_id"],
            "location_name": sensor_row["location_name"],
            "sensor_id": sensor_id,
            "parameter": m["parameter"]["name"],
            "value": m["value"],
            "unit": m["parameter"]["units"],
            "datetime_utc": m["period"]["datetimeFrom"]["utc"],
            "datetime_local": m["period"]["datetimeFrom"]["local"],
            "latitude": sensor_row["latitude"],
            "longitude": sensor_row["longitude"]
        })

openaq_df = pd.DataFrame(all_measurement_rows)
openaq_df.head()
```

```python
openaq_df.shape
```

```python
openaq_df.to_csv("../data/raw/openaq_delhi_sample.csv", index=False)
```

```python
weather_url = "https://api.open-meteo.com/v1/forecast"

weather_params = {
    "latitude": 28.6139,
    "longitude": 77.2090,
    "hourly": "temperature_2m,relative_humidity_2m,wind_speed_10m,wind_direction_10m,precipitation",
    "timezone": "Asia/Kolkata",
    "forecast_days": 1
}

weather_response = requests.get(weather_url, params=weather_params)

print(weather_response.status_code)
print(weather_response.url)
```

```python
weather_data = weather_response.json()
weather_data.keys()
```

```python
hourly = weather_data["hourly"]

weather_df = pd.DataFrame({
    "datetime_local": hourly["time"],
    "temperature_2m": hourly["temperature_2m"],
    "relative_humidity_2m": hourly["relative_humidity_2m"],
    "wind_speed_10m": hourly["wind_speed_10m"],
    "wind_direction_10m": hourly["wind_direction_10m"],
    "precipitation": hourly["precipitation"]
})

weather_df.head()
```

```python
weather_df.shape
```

```python
weather_df.to_csv("../data/raw/openmeteo_delhi_sample.csv", index=False)
```
