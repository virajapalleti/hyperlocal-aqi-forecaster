```python
nested_response = {
    "results": [
        {
            "station": "Anand Vihar",
            "pollutant": "PM2.5",
            "value": 180,
            "coordinates": {
                "latitude": 28.6469,
                "longitude": 77.3152
            }
        },
        {
            "station": "Rohini",
            "pollutant": "PM2.5",
            "value": 140,
            "coordinates": {
                "latitude": 28.7383,
                "longitude": 77.0822
            }
        },
        {
            "station": "Dwarka",
            "pollutant": "PM10",
            "value": 220,
            "coordinates": {
                "latitude": 28.5921,
                "longitude": 77.0460
            }
        }
    ]
}
first_record = nested_response["results"][0]
first_record
```

```python
nested_response["results"]
```

```python
nested_response["results"][1]["coordinates"]["latitude"]
```

```python
rows = []

for record in nested_response["results"]:
    row = {
        "station": record["station"],
        "pollutant": record["pollutant"],
        "value": record["value"],
        "latitude": record["coordinates"]["latitude"],
        "longitude": record["coordinates"]["longitude"]
    }
    rows.append(row)

rows
```

```python
nested_df = pd.DataFrame(rows)
nested_df
```

```python
url = "https://api.github.com/search/repositories"

params = {
    "q": "pandas language:python",
    "sort": "stars",
    "order": "desc",
    "per_page": 5
}

response = requests.get(url, params=params)
data = response.json()
```

```python
data.keys()
```

```python
data["items"][0].keys()
```

```python
repos = data["items"]

clean_rows = []

for repo in repos:
    row = {
        "name": repo["name"],
        "full_name": repo["full_name"],
        "stars": repo["stargazers_count"],
        "language": repo["language"],
        "url": repo["html_url"]
    }
    clean_rows.append(row)

clean_rows
```

```python
repos_df = pd.DataFrame(clean_rows)
repos_df
```
