```python
import pandas as pd
```

```python
aqi_df = pd.read_csv("../data/raw/openaq_delhi_sample.csv")
aqi_df.head()
```

```python
#standardize column names
aqi_df.columns = (
    aqi_df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

aqi_df.columns
```

```python
#now we're converting timestamp text into real date time objs to sort and filter easily for later
aqi_df["datetime_utc"] = pd.to_datetime(aqi_df["datetime_utc"])
aqi_df["datetime_local"] = pd.to_datetime(aqi_df["datetime_local"]
                                          )

aqi_df.info()
```

```python
#checking pollutant types

aqi_df["parameter"].unique()
```

```python
aqi_pm25_df = aqi_df[aqi_df["parameter"] == "pm25"].copy()
aqi_pm25_df.head()
#keeping only pm2.5 data for now, since it's the most commonly used pollutant for AQI calculation and forecasting. We can always add more pollutants later if needed.
```

```python
aqi_pm25_df.head()
```

```python
aqi_pm25_df = aqi_pm25_df.drop_duplicates()
```

```python
aqi_pm25_df.shape
```

```python
aqi_pm25_df.isna().sum()
```

```python
aqi_pm25_df[aqi_pm25_df["value"].isna()]
```

```python
aqi_pm25_df = aqi_pm25_df.dropna(subset=["value", "datetime_local", "location_name"])
```

```python
aqi_pm25_df = aqi_pm25_df[aqi_pm25_df["value"] >= 0]
#removing negative valuesv = cause they're impossible
```

```python
aqi_pm25_df = aqi_pm25_df.sort_values(["location_name", "datetime_local"])
```

```python
aqi_pm25_df = aqi_pm25_df.sort_values(["location_name", "datetime_local"])
```

```python
aqi_pm25_df.to_csv("../data/processed/clean_aqi_pm25.csv", index=False)
```
